```python
                            #DATA LOADING 

import pandas as pd

file_path = 'Cancer.xlsx'
data = pd.read_excel(file_path, sheet_name='Sheet1')
data

```

```python
                      #REMOVE OUTLIERS

#Identify numerical columns
numeric_data=data.select_dtypes(include='number')

#Calculate Q1 and Q3 and IQR
Q1=numeric_data.quantile(0.25)
Q3=numeric_data.quantile(0.75)
IQR=Q3-Q1

#Define outlier bounds
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR

#Create a mask for outliers (Use logical conditions to identify rows that contain outliers in any numerical column)
outlier_mask=~((numeric_data<lower_bound)|(numeric_data>upper_bound)).any(axis=1) #axis=1 confirms that we are working on rows

#Filter the Dataframe to remove outliers,keeping non-numerical columns
data_no_outliers=data[outlier_mask]

#Display the results
print("Original Dataframe:")
print(data)
print("\nDataFrame without outliers:")
print(data_no_outliers)
```

```python
#204 rows X 27 columns originally with possible outliers
#204 rows X 27 columns without outliers.
#Therefore no outliers have been found present in the dataframe.
```
